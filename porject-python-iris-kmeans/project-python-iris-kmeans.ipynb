```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBacked.figure_format='retina'
```

```python
# import and read iris dataset in csv
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/iris.csv')
```

```python
# count dataset row
df.species.value_counts()
```

```python
# show 10 sample
df.sample(10)
```

```python
# show 50 sample
df.sample(50)
```

### Kmean

```python
# import sciencki learn for use kmean
from sklearn.cluster import KMeans
```

```python
# use kmean
model=KMeans(n_clusters=3)
X=df[['petal_length','petal_width']]
model.fit(X)
model.cluster_centers_
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1', marker='D')
```

```python
# validate accurancy
df['cluster']=model.labels_ # cluster
pd.crosstab(df['species'], df['cluster']) # crosstab
```

```python
model=KMeans(n_clusters=3)
X=df[['sepal_length','sepal_width']]
model.fit(X)
model.cluster_centers_
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='.1', marker='D')
```

```python
df['cluster']=model.labels_ # cluster
pd.crosstab(df['species'], df['cluster']) # crosstab
```

### svm

```python
from sklearn import svm
```

```python
X = [[0, 0], [1, 1]]
y = [0, 1]
```

```python
clf = svm.SVC(gamma='scale')
```

```python
clf.fit(X, y)
```

```python
clf.predict([[2., 2.]])
```

```python
# get support vectors
clf.support_vectors_


# get indices of support vectors
clf.support_ 

# get number of support vectors for each class
clf.n_support_ 
```
